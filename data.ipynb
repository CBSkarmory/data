### Setup

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
from matplotlib import rc
import numpy as np
```

```python
# enables LaTeX
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

### Should you sit in the front row? 

```python
objects = ('0', r"$\forall n \in \mathbb{Z}^{+}$")
y_pos = np.arange(len(objects))
performance = [0,1]
 
plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('Probability you should')
plt.xlabel('Number of seats open in the front row')
plt.title('Should You Sit in the Front Row?')

plt.annotate('Source: studies that show this'
             , (0,0), (-30,-35), fontsize=7, 
             xycoords='axes fraction', textcoords='offset points',va='top')


plt.savefig('./exports/front_row.png')
plt.show()
```
